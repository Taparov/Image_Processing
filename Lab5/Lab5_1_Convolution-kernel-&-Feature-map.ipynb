````python
### START CODE HERE ###
first_conv = vgg16.features[0]
first_relu = vgg16.features[1]
first_conv_data = first_conv.weight.data
print(first_conv_data.shape)
for i in range(first_conv_data.shape[0]):
    print(f"Kernel: {i}\n*****************")
    for channel in range(3):
        print(f"Channel : {channel}")
        arr = first_conv_data[i,channel,:,:].cpu().numpy()
        print(arr)
        print(f"Min coefficients {arr.min()}")
    print("*****************")

### END CODE HERE ###
````
